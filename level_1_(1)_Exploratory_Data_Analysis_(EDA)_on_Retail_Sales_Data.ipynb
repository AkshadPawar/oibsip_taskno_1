```python
!pip install gdown

import pandas as pd
import gdown

# Download the file from Google Drive
file_id = '1n108end-wx9W7hW4PL7VIfNMN8eJRl3J' # extract the ID from the Google Drive link
url = f'https://drive.google.com/uc?id={file_id}'
output_file = 'sales_data.csv'  # Define a local file name
gdown.download(url, output_file, quiet=False)

# Load the dataset
file_path = output_file  # Update the file_path to the local file
data = pd.read_csv(file_path)

```

```python
# Basic Descriptive Statistics
descriptive_stats = data.describe()
print(descriptive_stats)

# Additional Statistics: Mode and median for numeric columns
numeric_columns = data.select_dtypes(include=['int64', 'float64']).columns

# Calculate mode for numeric columns
mode_values = data[numeric_columns].mode().iloc[0]

# Calculate median for numeric columns
median_values = data[numeric_columns].median()

print("Mode Values for Numeric Columns:\n", mode_values)
print("Median Values for Numeric Columns:\n", median_values)
```

```python
# Time Series: Sales trends over time
import matplotlib.pyplot as plt
daily_sales = data.groupby('Date').agg({'Total Amount': 'sum'}).reset_index()

# Plotting sales over time
plt.figure(figsize=(10, 6))
plt.plot(daily_sales['Date'], daily_sales['Total Amount'], label='Total Sales', color='blue')
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

```

```python
# Customer Age Distribution
plt.figure(figsize=(10,6))
plt.hist(data['Age'], bins=20, color='green', edgecolor='black')
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Gender Distribution
gender_counts = data['Gender'].value_counts()
gender_counts.plot(kind='bar', color='orange')
plt.title('Gender Distribution of Customers')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Product Category Distribution
product_category_counts = data['Product Category'].value_counts()

plt.figure(figsize=(12,6))
product_category_counts.plot(kind='bar', color='purple')
plt.title('Product Category Distribution')
plt.xlabel('Product Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

```

```python
# Total sales by product category
category_sales = data.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
category_sales.plot(kind='bar', color='teal')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Average spending by customer age
age_spending = data.groupby('Age')['Total Amount'].mean()

plt.figure(figsize=(10,6))
age_spending.plot(kind='line', color='brown')
plt.title('Average Spending by Customer Age')
plt.xlabel('Age')
plt.ylabel('Average Spending')
plt.show()
```
